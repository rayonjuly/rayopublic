```python
def normalize_column_name(df):
    import re
    from unidecode import unidecode
    result = []
    for col in df.columns:
        col = re.sub('(\n|\t)', '', col) # replace \n and \t with ''
        col = col.lower() # lower
        col = re.sub(' ', '_', col) # repace ' ' with '_'
        col = unidecode(col) # remove accent
        
        result.append(col)
        
    df.columns = result
        
    return df
```

# 1. Load data

```python
df = pd.read_excel(r'D:\MINHTL3\rayopublic\8. Extract info from papers\raw_data\Extracted Info.xlsx', sheet_name= 'Vietnamese')

to_replace = ['Không rõ', 'Không có thông tin', 'không có thông tin', 'không rõ', 'Inconnu', "Pas d'informations"]
df = normalize_column_name(df)
df = df[~(df['ten_file'] == 'Tên file')]
df = (
    df
    .dropna(subset = 'ten_file')
    .assign(stt =                               lambda df_: df_.ten_file.str.extract(r"^(\d+(?:\.\d+)*\.?)"))
    .assign(ten_file_khong_extension =          lambda df_: df_['ten_file'].str.removesuffix(r'.docx').str.removesuffix(r'.doc').str.removesuffix(r'.pdf').str.removesuffix(r'.PDF').str.removesuffix(r'.ppt'))
    .drop_duplicates(subset=['stt','ten_file_khong_extension'])
    .reset_index(drop=True)
    # .replace(to_replace, np.nan, inplace=True)
)
df
```
