```python
class Solution:
    def summarise(self, nums):
        result = [nums[0]]
        for i in nums[1:]:
            if i * result[-1] > 0:
                result[-1] += i
            else:
                result.append(i)
                
        return result
    
    def maxAbsoluteSum(self, nums) -> int:
        newnums = self.summarise(nums)
        result = 0
        for i in range(0, len(newnums)):
            for j in range(i, len(newnums)):
                s = abs(sum(newnums[i:j+1]))
                result = max(result, s)
        return result
        
s = Solution()
nums = [1,-3,2,3,-4]
s.summarise(nums)
s.maxAbsoluteSum(nums)
```
