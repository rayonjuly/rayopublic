```python
def longest_increasing_sequence(pairs):
    pairs.sort()  # Step 1: Sort by `a`, then `b`
    
    def binary_search(tails, target):
        """Find the leftmost index to replace target in LIS"""
        left, right = 0, len(tails) - 1
        while left <= right:
            mid = (left + right) // 2
            if tails[mid][1] >= target[1]:  # Ensure increasing `b`
                right = mid - 1
            else:
                left = mid + 1
        return left
    
    tails = []
    prev_index = {}  # Track previous elements for reconstructing path
    index_map = {}  # Map indices for backtracking
    
    for i, pair in enumerate(pairs):
        pos = binary_search(tails, pair)
        
        if pos < len(tails):
            tails[pos] = pair  # Replace to maintain smallest increasing sequence
        else:
            tails.append(pair)  # Extend LIS

        if pos > 0:
            prev_index[pair] = tails[pos - 1]  # Track previous element in sequence
        
        index_map[pair] = i  # Store index for reconstruction
    
    # Reconstruct the longest path
    longest_path = []
    last = tails[-1] if tails else None
    while last:
        longest_path.append(last)
        last = prev_index.get(last, None)
    
    return longest_path[::-1]  # Reverse to get correct order


# Example usage
# pairs = [(5, 24), (15, 25), (27, 40), (50, 60), (10, 30)]
# longest_path = longest_increasing_path(pairs)
# print(longest_path)
longest_increasing_sequence(tuple_list)
# tuple_list
```
